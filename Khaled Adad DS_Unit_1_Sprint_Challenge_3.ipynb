```python
contingency_table = pd.crosstab(df['A16'], df['A1'])
contingency_table
```

```python
chi_squared, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-Squared: {chi_squared}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}") 
print("Expected: \n", np.array(expected))
```

```python
p_value = stats.chi2.sf(chi_squared, dof)
print(f"P-value: {p_value}")
```

We accept the null hypothesis that A1 and A16 are independent

```python
contingency_table2 = pd.crosstab(
    df['A16'],
    df['A7'],
    margins = True)

contingency_table2
```

#### We reject the null hypothesis that A7 and A16 are independent
