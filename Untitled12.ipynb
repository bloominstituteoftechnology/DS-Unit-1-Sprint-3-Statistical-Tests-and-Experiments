```
import pandas as pd 
import scipy
from scipy import stats

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data', sep=',', header=None, skiprows=1, names = ["A1", "A2", "A3", "A4", "A5", "A6",
                                                                                                                                                    "A7", "A8", "A9", "A10", "A11", "A12",
                                                                                                                                                    "A13", "A14", "A15", "A16"])
data.dtypes

```

```
def unique_entries(column):
  return data.groupby(column).count()

unique_entries(data['A4'])
```

```
pd.crosstab(data.A3, data.A16, margins=True)
```

```
data = data.replace('-', 1)
data = data.replace('+', 0)

data[15] = pd.to_numeric(data[15])



approved = data.loc[data[15] == 1]
denied = data.loc[data[15] == 0]

scipy.stats.ttest_1samp(data[15], 0.5)
```
