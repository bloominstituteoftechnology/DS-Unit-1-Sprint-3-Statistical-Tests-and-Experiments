```
credit.head()
```

```
approval_by_sex = credit[['male','approved']]
approval_by_sex.head()
```

```
contingency_table = pd.crosstab(approval_by_sex['male'],
                                approval_by_sex['approved'],
                               margins = True)
contingency = pd.crosstab(approval_by_sex['male'],
                                approval_by_sex['approved'])

contingency_table
```

```
import scipy.stats as stats
chi_squared, p_value, dof, expected = stats.chi2_contingency(contingency)
print(f"Chi-Squared: {chi_squared}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}") 
print("Expected: \n", np.array(expected))
'''p value is .57 so I fail to reject the null that they are independant, 
which was expected based on the crosstab''' 
```

```
contingency=pd.crosstab(credit['prior_default'],credit['approved'])
print(contingency)
chi_squared, p_value, dof, expected = stats.chi2_contingency(contingency)
print(f"Chi-Squared: {chi_squared}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}") 
print("Expected: \n", np.array(expected))
'''Ok this one has a high chi^2 value and a very low p-value. 
I would reject the null hypothesis that they are independent'''
```

```
contingency=pd.crosstab(credit['employed'],credit['approved'])
print(contingency)
chi_squared, p_value, dof, expected = stats.chi2_contingency(contingency)
print(f"Chi-Squared: {chi_squared}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}") 
print("Expected: \n", np.array(expected))
'''Another one where I would reject the null hypothesis of independence
based on the significant p-value and chi^2 statistic.'''
```

```
debt= credit[['debt','approved']]

approved = debt.set_index(['approved']).loc[1]
unapproved = debt.set_index(['approved']).loc[0]
stats.ttest_ind(approved.debt,unapproved.debt, nan_policy='omit')
```

```
score= credit[['credit_score','approved']]

approved = score.set_index(['approved']).loc[1]
unapproved = score.set_index(['approved']).loc[0]
stats.ttest_ind(approved['credit_score'],unapproved['credit_score'] , nan_policy='omit')
```
